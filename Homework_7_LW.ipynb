3.Remove any column deemed unnecessary. For the columns contains text values, please change the case to lower.
NOTE: I used the indexes in the csv as the dataframe index so I would not have an unnecessary column

```python
marketing_df['job']=marketing_df['job'].str.lower()
marketing_df['marital']=marketing_df['marital'].str.lower()
marketing_df['education']=marketing_df['education'].str.lower()
marketing_df['default']=marketing_df['default'].str.lower()
marketing_df['housing']=marketing_df['housing'].str.lower()
marketing_df['loan']=marketing_df['loan'].str.lower()
marketing_df['contact']=marketing_df['contact'].str.lower()
marketing_df['month']=marketing_df['month'].str.lower()
marketing_df['poutcome']=marketing_df['poutcome'].str.lower()
marketing_df['deposit']=marketing_df['deposit'].str.lower()
marketing_df.head()
```

5. What is the age group of the people who took the service of term deposit from the bank? Visualize the outcome using plots.

```python
min_age =marketing_df[marketing_df['deposit']=='yes']['age'].min()
max_age= marketing_df[marketing_df['deposit']=='yes']['age'].max()
print("The people who took the service of term deposit from the bank are between %d and %d" % (min_age, max_age))
```

```python
marketing_df[marketing_df['deposit']=='yes']['age'].plot.hist(column=["age"])
```
