```python
#imports
import pandas as pd
```

```python
#read in course sections
df1 = pd.read_excel('StudentsDB.xlsx', sheet_name='COURSE_SECTION1')
df2 = pd.read_excel('StudentsDB.xlsx', sheet_name='COURSE_SECTION2')
df1 =df1.dropna()
df1
```

```python
section_df =pd.concat([df1,df2])
```

```python
#read in students
student_df = pd.read_excel('StudentsDB.xlsx',
                      sheet_name ='STUDENT')
```

```python
#read in enrollment
enrollment_df =  pd.read_excel('StudentsDB.xlsx',
                      sheet_name ='ENROLLMENT')
enrollment_df
```

```python
enrollment_df['GRADE'] =enrollment_df['GRADE'].fillna('In-Progress')
enrollment_df
```

```python
#read in courses
course = pd.read_excel('StudentsDB.xlsx', 'COURSE')
course = course.dropna()
course
```

```python
#merge all frames together 
new_df=student_df.merge(enrollment_df, left_on='S_ID', right_on='S_ID').merge(section_df, left_on='C_SEC_ID', right_on='C_SEC_ID').merge(course,  left_on='COURSE_ID', right_on='COURSE_ID')
```
